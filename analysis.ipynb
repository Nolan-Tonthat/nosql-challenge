### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
London_query = {"LocalAuthorityName":{"$regex":"London"},"RatingValue":{"$gte":4}}

# Use count_documents to display the number of documents in the result
establishments.count_documents(London_query)
```

```python
# Display the first document in the results using pprint
pprint(establishments.find(London_query)[0])
```

```python
#Convert query cursor into list
london_list = list(establishments.find(London_query))
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(london_list)
# Display the number of rows in the DataFrame
print(len(london_df))
```

```python
# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
#Find latitude and longitude of Penang Flavours
query = {"BusinessName":"Penang Flavours"}
PenangF = establishments.find_one(query)

lat = PenangF["geocode"]["latitude"]
lon = PenangF["geocode"]["longitude"]

print(lat)
print(lon)
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
degree_search = 0.01

# Rating value must equal 5
rad_query = {"RatingValue":5, "geocode.latitude": {"$gte":(lat-degree_search),"$lte":(lat+degree_search)},
                "geocode.longitude": {"$gte":(lon-degree_search),"$lte":(lon+degree_search)}}

# Sort by hygiene score
sort = "scores.Hygiene"

# Print the results
top5_result = list(establishments.find(rad_query).sort(sort,1).limit(5))
pprint(top5_result)
```

```python
# Convert result to Pandas DataFrame
top5_df = pd.DataFrame(top5_result)
top5_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match_zero = {"$match":{"scores.Hygiene": 0}}

# 2. Groups the matches by Local Authority
group_by =  {"$group":{"_id":"$LocalAuthorityName", "count":{"$sum":1}}}

# 3. Sorts the matches from highest to lowest
sort_by = {"$sort":{"count":-1}}


zeroH_result = list(establishments.aggregate([match_zero,group_by,sort_by])) #aggregation must be turned into list for print function to work (aggregation returns cursor object)

# Print the number of documents in the result
print(len(zeroH_result))

# Print the first 10 results
for x in range(10):
    pprint(zeroH_result)
```

```python
# Convert the result to a Pandas DataFrame
zero_hygiene_df = pd.DataFrame(zeroH_result)

# Display the number of rows in the DataFrame
print(len(zero_hygiene_df))
```

```python
# Display the first 10 rows of the DataFrame
zero_hygiene_df.head(10)
```
